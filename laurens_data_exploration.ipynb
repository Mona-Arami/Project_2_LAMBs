```python
crime_df = pd.read_csv('crime.csv')
crime_df.head()
```

```python
crime_1_df = crime_df[crime_df.OFFENSE_CATEGORY_ID != 'traffic-accident']
crime_1_df.head()
```

```python
pd.write_csv()
```

```python
# denver crime data cleaned up by filtering columns
crime_2_df = crime_df[['INCIDENT_ID', 'OFFENSE_ID', 'OFFENSE_CODE', 'OFFENSE_TYPE_ID', 'OFFENSE_CATEGORY_ID', 'REPORTED_DATE', 'GEO_LON', 'GEO_LAT', 'NEIGHBORHOOD_ID']]
crime_2_df.head()
```

```python
# number of crimes in denver per category (not sure difference between incident id and offense id)
crimes_by_category_df = crime_2_df.groupby('OFFENSE_CATEGORY_ID')[['INCIDENT_ID', 'OFFENSE_ID']].nunique()
crimes_by_category_df
```

```python
# number of crimes per neighborhood (sorted least reported crime to most)
crimes_by_neighborhood_df = crime_2_df.groupby('NEIGHBORHOOD_ID')[['INCIDENT_ID', 'OFFENSE_ID']].nunique()
crimes_by_neighborhood_df.sort_values('OFFENSE_ID').head()
```

```python
# number of crimes per neighborhood (viewing 5 neighborhoods with highest numbers of reported crimes)
crimes_by_neighborhood_df.sort_values('OFFENSE_ID').tail()
```
